```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("taj.jpg.webp")

if image is None:
    print("Image not found")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
M = np.float32([[1,0,100],[0,1,50]])
translated = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))

image_rgb = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Translated Image")
plt.axis('off')
plt.show()
```

```python
scaled = cv2.resize(image, None, fx=1.5, fy=1.5)
image_rgb = cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Scaled Image")
plt.axis('off')
plt.show()
```

```python
rows, cols, _ = image.shape
M = np.float32([[1,0.5,0],[0,1,0]])
sheared = cv2.warpAffine(image, M, (cols, rows))
image_rgb = cv2.cvtColor(sheared, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Sheared Image")
plt.axis('off')
plt.show()
```

```python
reflected = cv2.flip(image,1)

image_rgb = cv2.cvtColor(reflected, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Reflected Image")
plt.axis('off')
plt.show()
```

```python
(h,w) = image.shape[:2]
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center,45,1)
rotated = cv2.warpAffine(image,M,(w,h))
image_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Rotated Image")
plt.axis('off')
plt.show()
```
